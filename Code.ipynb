```python
#Import the function to calculate the average
import math
```

```python
#List of the average values necessary to create bar
value_7b=[]
#List of the average error values necessary to create error bar
value_error_7b=[]

#Definicion of function calculating the average value
def average(x):
    value = sum(x)
    return value / float(len(x))
    average(WT_KON_7b)
    average(WT_KON_7b_error)
    average(WT_RS_7b)
    average(WT_RS_7b_error)
    average(NETKO_KON_7b)
    average(NETKO_KON_7b_error)
    average(NETKO_RS_7b)
    average(NETKO_RS_7b_error)
    average(SWR_KON_7b)
    average(SWR_KON_7b_error)
    average(SWR_RS_7b)
    average(SWR_RS_7b_error)

#Adding the values to one list 
value_7b.append(average(WT_KON_7b))
value_7b.append(average(WT_RS_7b))
value_7b.append(average(NETKO_KON_7b))
value_7b.append(average(NETKO_RS_7b))
value_7b.append(average(SWR_KON_7b))
value_7b.append(average(SWR_RS_7b))
```

```python
#Adding the error values to one list  
value_error_7b.append(average(WT_KON_7b_error))
value_error_7b.append(average(WT_RS_7b_error))
value_error_7b.append(average(NETKO_KON_7b_error))
value_error_7b.append(average(NETKO_RS_7b_error))
value_error_7b.append(average(SWR_KON_7b_error))
value_error_7b.append(average(SWR_RS_7b_error))
```
